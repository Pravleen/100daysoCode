# STEP 3 HANDLE MISSING DATA 

```python
data.dropna(subset=["Price"],inplace=True,axis=0)
```

```python
data
```

```python
mean=data["Wheel-base"].mean()
mode=data["Stroke"].mode().iloc[0]
mode1=data["Bore"].mode().iloc[0]
mode2=data["Hp"].mode().iloc[0]
mode3=data["Peak-rpm"].mode().iloc[0]
```

```python
data['Normalised-loses'].replace('?',mean,inplace=True)
data['Price'].replace('?',mean,inplace=True)
data['Stroke'].replace('?',mode,inplace=True)
data['Bore'].replace('?',mode1,inplace=True)
data['Hp'].replace('?',mode2,inplace=True)
data['Peak-rpm'].replace('?',mode3,inplace=True)
```

```python
data.iloc[57:61,:]
```

```python
s=data.isnull().sum(axis = 0)
s
```

```python
data["Price"]=data["Price"].astype('float')
data["Stroke"]=data["Stroke"].astype('float')
data["Bore"]=data["Bore"].astype('float')
```
