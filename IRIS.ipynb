```python
#import dependencies
import pandas as pd
from matplotlib import pylab
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("iris.data",header=None)
```

```python
df
```

```python
headers=["Sepel_length","Sepal_width","Petal_length","Petal_width","Class"]
df.columns=headers
```

```python
df["Sepel_length+new"]=df["Sepel_length"]/df["Sepel_length"].max()
```

```python
df.head()
```

```python
plt.hist(df["Sepel_length"])
plt.xlabel("Sepel length")
plt.ylabel("Count")
plt.title("Sepal length histogram")
plt.show()
```

```python
bins=np.linspace(df["Sepel_length"].min(),df["Sepel_length"].max(),4)
group_name=["Small","Medium","Large"]
```

```python
df["Sepel_length_binned"]=pd.cut(df["Sepel_length"],bins,labels=group_name,include_lowest=True)
```

```python
df[["Sepel_length","Sepel_length_binned"]]
```

```python
df["Sepel_length_binned"].value_counts()
```

```python
plt.hist(df["Sepel_length_binned"],color="red",bins=3)
plt.xlabel("Sepel length")
plt.ylabel("Count")
plt.title("Sepal length binned histogram")
plt.show()
```

```python
dummy_variable_1=pd.get_dummies(df["Class"])
dummy_variable_1.head()
```

```python
df=pd.concat([df,dummy_variable_1],axis=1)
```
